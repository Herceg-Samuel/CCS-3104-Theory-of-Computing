# Tourism Itinerary Parser using Context-Free Grammar (CFG)

## Theory of Computing Assignment

This notebook implements a Context-Free Grammar (CFG) parser for a tourism itinerary language using Python and NLTK.

The language supports:

- Planning trips
- Adding activities
- Transport selection
- Accommodation selection
- Date specification

The parser checks whether itinerary commands conform to the grammar before they are processed by a tourism booking system.



!pip install nltk

```python
import nltk

from nltk import CFG
from nltk.parse import ChartParser
```

```python
grammar = CFG.fromstring("""
S -> Command ';'
Command -> PlanTrip | AddActivity

PlanTrip -> 'PLAN' 'trip' 'to' Location 'from' Date 'to' Date 'activity' Activity 'transport' Transport 'hotel' Hotel
AddActivity -> 'ADD' 'activity' Activity 'on' Date

Location -> 'MaasaiMara' | 'Diani' | 'MtKenya' | 'Samburu' | 'LakeNakuru' | 'Amboseli'
Activity -> 'safari' | 'beach' | 'hiking' | 'birdwatching'
Transport -> 'van' | 'flight' | 'train' | 'landcruiser'
Hotel -> 'MaraLodge' | 'BeachResort' | 'MountainCamp' | 'LakeView'
Date -> '2026-08-01' | '2026-08-05' | '2026-09-10' | '2026-09-14' | '2026-08-03'
""")
```

```python
parser = ChartParser(grammar)
```

```python
tokens = [
    "PLAN",
    "trip",
    "to",
    "MaasaiMara",
    "from",
    "2026-08-01",
    "to",
    "2026-08-05",
    "activity",
    "safari",
    "transport",
    "van",
    "hotel",
    "MaraLodge",
    ";"
]
```

```python
trees = list(parser.parse(tokens))
```

```python
if trees:
    print("Valid itinerary!\n")
else:
    print("Invalid itinerary.")
```

```python
for tree in trees:
    print(tree)
```

```python
for tree in trees:
    tree.pretty_print()
```

```python
def tokenize(sentence):
    sentence = sentence.replace(";", " ;")
    return sentence.split()
```

```python
def parse_sentence(sentence):

    tokens = tokenize(sentence)

    print("=" * 70)
    print("INPUT:")
    print(sentence)

    print("\nTOKENS:")
    print(tokens)

    try:
        trees = list(parser.parse(tokens))

        if trees:
            print("\nRESULT: VALID STATEMENT\n")

            for i, tree in enumerate(trees, 1):
                print(f"Parse Tree {i}")
                print(tree)
                print()

                tree.pretty_print()

        else:
            print("\nRESULT: INVALID STATEMENT")

    except ValueError as e:
        print("\nLEXICAL ERROR")
        print(e)
```

```python
test_cases = [
    "PLAN trip to MaasaiMara from 2026-08-01 to 2026-08-05 activity safari transport van hotel MaraLodge;",
    "PLAN trip to Diani from 2026-09-10 to 2026-09-14 activity beach transport flight hotel BeachResort;",
    "ADD activity hiking on 2026-08-03;",
    "PLAN MaasaiMara trip safari date;"
]

for case in test_cases:
    parse_sentence(case)
```
