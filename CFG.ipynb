```python
import nltk
from nltk.tree import Tree
from nltk.draw import tree
from nltk import CFG
from nltk.grammar import *
```

```python
grammar_c = CFG.fromstring("""
S -> T | U
T -> 'a' 'a' T 'b' | 'a' T | 'a'
U -> 'a' C | C
C -> 'a' 'a' C 'b' | C 'b' | 'b'
""")
```

```python
sent_c = list("aaab")
rd_parser = nltk.ChartParser(grammar_c)
for t in rd_parser.parse(sent_c):
    print(t)
```

```python
grammar_d = CFG.fromstring("""
S -> 'a' S 'b' 'b' | 'a' S 'b' 'b' 'b' |
""")
```

```python
sent_d = list("aabbbb")
rd_parser = nltk.RecursiveDescentParser(grammar_d)
for t in rd_parser.parse(sent_d):
    print(t)
```

```python
grammar_e = CFG.fromstring("""
S -> T 'a' T | T 'b' T
T -> 'a' T 'b' T | 'b' T 'a' T |
""")
```

```python
grammar_f = CFG.fromstring("""
S -> 'a' S 'b' S | 'a' S |
""")
```

```python
sent_f = list("aabb")
parser = nltk.ChartParser(grammar_f)
trees_f = list(parser.parse(sent_f))
print(trees_f[0])

for t in parser.parse(sent_f):
    print(t)
```

```python
grammar_g = CFG.fromstring("""
S -> 'a' E
E -> 'a' E 'a' E 'b' E | 'a' E 'b' E 'a' E | 'b' E 'a' E 'a' E |
""")

grammar_g.start()

grammar_g.productions()
```

```python
examples = {
    "c": (grammar_c, "aaab"),
    "d": (grammar_d, "aabbbb"),
    "e": (grammar_e, "aab"),
    "f": (grammar_f, "aabb"),
    "g": (grammar_g, "aaab"),
}

for name, (g, s) in examples.items():
    print(f"Grammar ({name}) parsing '{s}' :")
    rd_parser = nltk.ChartParser(g)
    for t in rd_parser.parse(list(s)):
        print(t)

sr_parser = nltk.ShiftReduceParser(grammar_e)
sent = list("aab")
for t in sr_parser.parse(sent):
    print(t)
```
